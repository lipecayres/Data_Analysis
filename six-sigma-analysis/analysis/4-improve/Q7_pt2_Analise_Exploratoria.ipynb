##**Análise descritiva e exploratória**

**Bibliotecas utilizadas**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Importar dados pre processados**

```python
dataset = pd.read_csv("Dados_pre_processados.csv")
```

**Gráficos**

```python
sns.lmplot('cookie-length', 'cookie-width', data = dataset, fit_reg = False,
           hue = "status", scatter_kws = {"marker": "D", "s": 50})

plt.title('Comprimento do Biscoito vs Largura do Biscoito')
plt.xlabel("Comprimento do biscoito")
plt.ylabel("Largura do biscoito")

plt.show()
```

```python
sns.lmplot('cookie-thickness', 'cookie-length', data = dataset, fit_reg = False,
           hue = "status", scatter_kws = {"marker": "D", "s": 50})

plt.title('Espessura do Biscoito vs Largura do biscoito')
plt.xlabel("Espessura do biscoito")
plt.ylabel("Largura do biscoito")

plt.show()
```

```python
sns.lmplot('weight-before', 'cookie-thickness', data = dataset, fit_reg = False,
           hue = "status", scatter_kws = {"marker": "D", "s": 50})

plt.title('Peso cru do biscoito vs Espessura do biscoito')
plt.xlabel("Peso cru do biscoito")
plt.ylabel("Espessura do biscoito")

plt.show()
```

```python
sns.lmplot('weight-before', 'cookie-length', data = dataset, fit_reg = False,
           hue = "status", scatter_kws = {"marker": "D", "s": 50})

plt.title('Peso cru do biscoito vs Largura do biscoito')
plt.xlabel("Peso cru do biscoito")
plt.ylabel("Largura do biscoito")

plt.show()
```
