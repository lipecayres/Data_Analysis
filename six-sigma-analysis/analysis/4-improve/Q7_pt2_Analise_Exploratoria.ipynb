**Bibliotecas utilizadas**

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Importar dados pre processados**

```
dataset = pd.read_csv("Dados_pre_processados.csv")
```

**Gráficos**

```
sns.lmplot('sepal-length', 'sepal-width', data = dataset, fit_reg = False,
           hue = "class", scatter_kws = {"marker": "D", "s": 50})

plt.title('Comprimento da Sépala vs Largura da Sépala')
plt.xlabel("Comprimento da Sépala")
plt.ylabel("Largura da Sépala")

plt.show()
```

```
sns.lmplot('petal-length', 'petal-width', data = dataset, fit_reg = False,
           hue = "class", scatter_kws = {"marker": "D", "s": 50})

plt.title('Comprimento da Pétala vs Largura da Pétala')
plt.xlabel("Comprimento da Pétala")
plt.ylabel("Largura da Pétala")

plt.show()
```

```
sns.lmplot('sepal-length', 'sepal-width', data = dataset, fit_reg = False,
           hue = "class", scatter_kws = {"marker": "D", "s": 50})

plt.title('Comprimento da Sépala vs Comprimento da Pétala')
plt.xlabel("Comprimento da Sépala")
plt.ylabel("Comprimento da Pétala")

plt.show()
```

```
sns.lmplot('sepal-width', 'petal-width', data = dataset, fit_reg = False,
           hue = "class", scatter_kws = {"marker": "D", "s": 50})

plt.title('Largura da Sépala vs Largura da Pétala')
plt.xlabel("Largura da Sépala")
plt.ylabel("Largura da Pétala")

plt.show()
```
